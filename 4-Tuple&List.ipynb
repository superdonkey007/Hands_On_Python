## **元组和列表**

### **元组——不可修改**

```python
tuple1=('研究生一班','研究生二班',1120193565,3220241343)
# 元组的定义——建
# 1.只需要在小括号()中添加元素，并使用逗号隔开即可。
# 2.用格式转化tuple(seq)
# 3.需要注意的是，python中tuple2=tuple1这样的操作是浅拷贝，二者指向同一个元素，如果要新建一个tuple2，必须使用tuple()
tuple2=tuple(tuple1[1:])
print(tuple2)


#元组的删除——删除
#元组的单个元素不允许删除，整个元组用del语句删除
del tuple2
# print (f"删除后的元组 tup :{tuple2} ")
#删除后再print会报错not defined


# 元组的访问——查
# 元组的索引访问和字符串相同，用单下标访问元素，用多下标表示切片
#如tuple1，正向索引为0，1，2，3，反向索引为-4，-3，-2，-1
print(tuple1[1])
print(tuple1[::-1])
print(tuple1[:-1:])


#元组的运算
#长度运算len()
print(len(tuple1))
#拼接和复制：+和*
a = (1, 2, 3)
b = (4, 5, 6)
print(a*3+b*3)
print((a+b)*3)
#成员判断 in
print(1120193565 in tuple1)
#迭代
for x in tuple1:
    print(x)


#元组的常用函数

#内置函数
# len(my_tuple): 返回元组中的元素数量。
# max(my_tuple): 返回元组中的最大元素。
# min(my_tuple): 返回元组中的最小元素。
# sum(my_tuple): 返回元组中所有元素的总和（如果元组内是数字的话）。
# tuple(iterable): 可以将其他可迭代对象转换为元组。
# sorted(my_tuple): 返回排序后的列表（元组本身不会改变）。
# sorted(iterable, key=None, reverse=False)
# iterable -- 可迭代对象。
list1=[(2,'c'),(3,'a'),(1,'d'),(4,'b')]
# key -- 主要是用来取进行比较的元素的函数，只有一个参数，具体的函数的参数就是取自于可迭代对象中，指定可迭代对象中的一个元素来进行排序。
list2=sorted(list1,key=lambda x:x[0])
list3=sorted(list1,key=lambda x:x[1])
print(list2)
print(list3)
list1=['1',2,2,'C','e']
# cmp -- 比较的函数，这个具有两个参数，参数的值都是从可迭代对象中取出，此函数必须遵守的规则为，大于则返回1，小于则返回-1，等于则返回0。
# python3不再支持 cmp 参数，而是使用 key 参数结合 functools.cmp_to_key() 来实现自定义排序。
from functools import cmp_to_key
list4=sorted(list1,key=cmp_to_key(lambda x,y:(ascii(x)>ascii(y))))  
print(list4)
# key参数设置为一个复杂的函数，也可以实现复杂的自定义排序，返回值设置一个元组，依次按值排序
list1=[(2,'c'),(2,'a'),(1,'d'),(4,'b')]
list5=sorted(list1,key=lambda x:(x[0],x[1]))
print(list5)
# reverse -- 排序规则，reverse = True 降序 ， reverse = False 升序（默认）。


# 方法
# count(x): 返回 x 在元组中出现的次数。
print(tuple1.count(1120193565))
# index(x): 返回 x 在元组中的索引，如果不存在则抛出异常。
print(tuple1.index(3220241343))
```

### **列表**

```python
# 列表的定义——建

# 1.只需要在中括号[]中添加元素，并使用逗号隔开即可。
# 2.用格式转化list(seq)
list1 = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
list2 = list(list1)
print(list1)
print(list2)

# 列表元素的插入——增

#list.append(obj)方法给列表末尾增添一个元素obj
#list.insert(index, obj)给index位置插入obj
#list.extend(seq) 用可迭代序列在列表末尾一次性追加另一个序列中的多个值,没有返回值,直接在原列表上操作

# 列表元素删除——删：

# # 按序号删除
# del语句可以按序号删除单个元素，也可以用切片的形式删除连续的几个元素
# 1.如果切片值的范围超出，比如对这个list1 del list1[2:1000],不会抛出错误，把2即之后的元素全部删除
# 2.如果步长为负数，就不会删除
del list1[1]
print(list1)
del list1[1:3]
print(list1)
# list.pop([index=-1])，默认最后一个元素，删除后返回该元素的值,如果序号不存在抛出一个indexerror
list1.pop(0)
print(list1)
# 按值删除list.remove(obj)，如果不存在会抛出一个valueerror
list1.remove('f')
print(list1)
# 列表清空(只剩下一个空列表)
list1.clear()
print(list1)
# 整个列表用del语句(在内存中)删除
del list1


# 列表的访问——查

list1 = ['a', 'b', 'c', 'd']
# 列表的索引访问和字符串相同，用单下标访问元素，用多下标表示切片
# 如list1，正向索引为0，1，2，3，反向索引为-4，-3，-2，-1
print(list1[1])
print(list1[::-1])
print(list1[:-1:])

# 列表的更新——修改：用单个和切片索引修改

list1[1] = 'p'
print(list1)
list1[1:3] = ['t', 'm']
print(list1)

# list常用函数和方法

# 内置函数
# 1	len(list) 列表元素个数
# 2	max(list) 返回列表元素最大值
# 3	min(list) 返回列表元素最小值
# 4	list(seq) 将序列转换为列表
# 5 sum(list) 返回元组中所有元素的总和（如果列表内是数字的话）。
# 6 sorted(list) 返回排序后的列表（列表本身不会改变）。

# 常用方法
# 1	list.append(obj)在列表末尾添加新的对象
# 2	list.count(obj)统计某个元素在列表中出现的次数
# 3	list.extend(seq)在列表末尾一次性追加另一个序列中的多个值（用新列表扩展原来的列表）
# 4	list.index(obj)从列表中找出某个值第一个匹配项的索引位置
# 5	list.insert(index, obj)将对象插入列表
# 6	list.pop([index=-1])移除列表中的一个元素（默认最后一个元素），并且返回该元素的值
# 7	list.remove(obj)移除列表中某个值的第一个匹配项
# 8	list.reverse()反向列表中元素
# 9	list.sort(key=None, reverse=False)对原列表进行排序
# 10 list.clear()清空列表
# 11 list.copy()复制列表
```
