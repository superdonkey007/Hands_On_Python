## 集合和字典
#### Python中的集合和字典都是以哈希表实现的,意味着:
* 集合中的元素和字典中的键都是不可修改的,只能是数字,元组和字符串,不能是列表,集合和字典.
* In运算符的时间复杂度为O(1)

### 集合——无序 不重复元素序列

```python
# 集合的创建-建

# 可以使用大括号{}创建集合,元素之间用逗号','分隔,或者也可以使用set(seq)函数创建集合。
# 空集合必须用set()创建,因为{}是创建新字典
set1 = {1, 2, 3, 4}
set2 = set([6, 7, 8, 9])
obj = {}
print(type(obj))


# 集合元素的添加-增

# set.add(obj) 添加单个元素
set1.add(5)
# set.update(seq)
set1.update(set2)  # 添加可迭代对象中的所有元素
print(set1)
set2.update([1, 2, 3, 4, 5])
print(set2)


# 集合元素的删除-删

# set.remove(obj) 删除单个元素,如果元素不存在报错keyerror
set2.remove(9)
print(set2)
# set.discard(obj) 删除单个元素,如果元素不存在不报错
set2.discard(9)
print(set2)
# set.pop() 随机删除集合中的一个元素


# 集合元素的查看-查

#集合是可迭代的,可以直接输出或是通过for迭代输出
# 集合内元素本质上是无序的,但现代Python版本中的集合可能会表现出一定的稳定性
# 集合和字典迭代时会按照插入顺序进行迭代,无序表现在无法通过下标和切片进行访问
for i in set1:
    print(i)


# 集合的修改-改

# 集合内元素只有增删,没有修改
# 集合之间有交,并,差等运算
print(set1 - set2)  # 差集
print(set1 | set2)  # 并集
print(set1 & set2)  # 交
print(set1 ^ set2)  # 不同时存在的元素


# 集合的常用函数

# 在原集合上操作的函数：
# 1 add(obj) — 向集合中添加一个元素 obj。
# 2 clear() — 清空集合中的所有元素。
# 3 difference_update(set, ...) — 从原集合中移除出现在其他集合中的所有元素。
# 4 discard(obj) — 如果元素存在于集合中，则将其移除。
# 5 intersection_update(set, ...) — 更新原集合，使其只保留与其他集合的公共元素。
# 6 pop() — 随机移除并返回一个元素。
# 7 remove(obj) — 移除指定元素 obj。
# 8 symmetric_difference_update(set) — 更新原集合，使其只包含两个集合中不同的元素。
# 9 update(set, ...) — 更新原集合，使其包含所有输入集合中的所有元素。

# 新建一个集合的函数：
# 1 copy() — 返回当前集合的一个浅拷贝。
# 2 difference(set, ...) — 返回一个新的集合，包含调用者中有但在其他集合中没有的元素。
# 3 intersection(set, ...) — 返回一个新的集合，包含所有输入集合中的公共元素。
# 4 isdisjoint(set) — 如果两个集合没有交集则返回 True。
# 5 issubset(set) — 如果调用者中的所有元素都包含在其他集合中，则返回 True。
# 6 issuperset(set) — 如果其他集合中的所有元素都包含在调用者中，则返回 True。
# 7 symmetric_difference(set) — 返回一个新的集合，包含两个集合中不同的元素。
# 8 union(set, ...) — 返回一个新的集合，包含所有输入集合中的所有元素。
print(set1.union({'a','b'}))

 
```

### 字典-无序的键(key):值(value)的集合。

```python
# 字典的创建-建

# 字典可以使用大括号 `{}` 创建，键值对之间用逗号 `,` 分隔，或者使用 `dict()` 构造函数创建。
# 
# 使用大括号创建字典
dict1 = {'name': 'Alice', 'age': 30}
print(dict1)
# 使用 dict(seq) 函数创建字典,seq元素必须是一个key, value序列,这个序列可以是二元元组,二元列表,二元集合,字典的键值对等
ls=[{1,'a'},{2,'b'}]
dict2 = dict(ls)
print(dict2)
# 空字典
empty_dict = {}
print(type(empty_dict))  # 输出：<class 'dict'>


# ### 字典元素的添加（增）

# 字典可以通过指定键并赋值的方式添加元素，也可以使用 `update()` 方法批量添加。
# 直接赋值添加元素
dict1['job'] = 'Engineer'
# # 使用 update() 方法添加多个键值对
dict1.update(dict2)
print(dict1)

# 字典元素的删除（删）

# 字典可以通过键来删除元素，如果键不存在，可以选择抛出异常或不抛出异常。

# 使用del和pop()方法删除元素，如果键不存在则抛出KeyError,pop()返回value值
del dict1[1]
print(dict1)
value = dict1.pop('age')
print(value)

# 使用 popitem() 方法删除最后一个插入的键值对（返回键值对）
last_key_value = dict1.popitem()
print(last_key_value)


# 字典元素的查看-查

# 字典可以通过键来访问值，也可以通过迭代遍历所有键值对。

# 访问字典中的值
print(dict1['name'])
# 直接遍历字典得到的实际上是keys
for key in dict1:
    print(key)
# 遍历字典中的所有键值对
for key, value in dict1.items():
    print(f"{key}: {value}")
# 遍历字典中的所有键
for key in dict1.keys():
    print(key)
# 遍历字典中的所有值
for value in dict1.values():
    print(value)


# 字典的修改-改
# 字典可以通过键来修改对应的值，也可以使用 `update()` 方法批量修改。

# 修改字典中的值
dict1['name'] = 'Charlie'
print(dict1)

# 使用 update(seq) 方法批量修改字典中的值,seq元素必须是一个key, value序列,这个序列可以是二元元组,二元列表,二元集合,字典的键值对等
dict1.update({'name': 'Diana', 'age': 35})
dict1.update(ls)
print(dict1)


# 字典的常用函数

#在原字典上操作的函数：
# 1. `clear()` — 清空字典中的所有键值对。
# 2. `get(key[, default])` — 返回指定键的值，如果键不存在则返回 `default`（默认为 `None`）。
# 3. `items()` — 返回字典中键值对的视图。
# 4. `keys()` — 返回字典中所有键的视图。
# 5. `pop(key[, default])` — 删除指定键并返回其值，如果键不存在则返回 `default`（默认抛出 `KeyError`）。
# 6. `popitem()` — 删除并返回最后一个插入的键值对。
# 7. `setdefault(key[, default])` — 如果键在字典中存在，则返回对应的值；否则插入键值对并返回 `default`。
# 8. `update([other])` — 更新字典，添加或更新来自 `other` 的键值对。
# 9. `values()` — 返回字典中所有值的视图。

# 新建一个字典的函数：
# 1. `fromkeys(seq[, v])` — 创建一个新字典，其中包含 `seq` 中的键，默认值为 `v`（默认为 `None`）。
# 2. `copy()` — 返回当前字典的一个浅拷贝。
# 3. `dict()` — 根据提供的键值对创建一个新的字典。
```
