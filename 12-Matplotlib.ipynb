## Matplotlib
Matplotlib 是一个强大的绘图库,用于创建各种类型的静态、动态和交互式图表.

Pyplot 是 Matplotlib 的子库,提供了和 MATLAB 类似的绘图 API.我们一般使用这个子库来创建图表.

### 包导入

```python
import matplotlib.pyplot as plt
# plot():用于绘制线图和散点图
# scatter():用于绘制散点图
# bar():用于绘制垂直条形图和水平条形图
# hist():用于绘制直方图
# pie():用于绘制饼图
# imshow():用于绘制图像
# subplots():用于创建子图
import numpy as np
```

### Pyplot作图工作流程
1. 用plot.fiture( )创建一个绘图对象

2. 用plt.{chart}( )传入数据和具体要绘制的图表

3. 用plt.{setter}( )传入调节图表信息和布局的参数

4. 用plt.show( )显示图表或者用plt.savefig()保存图表

5. 用plt.close( )关闭图表

下面是一段示例,看完这个示例后我将逐个介绍这几个阶段

```python
# 示例数据:一个月内每天的销售量
days = list(range(1, 31))  # 1到30天
sales = [250, 300, 280, 310, 290, 330, 340, 360, 300, 310,
         280, 350, 360, 390, 370, 400, 420, 430, 410, 390,
         400, 420, 430, 440, 450, 460, 480, 500, 490, 520]

# 创建图形对象
plt.figure(figsize=(10, 6))

# 绘制柱状图
plt.bar(days, sales, color='lightblue', alpha=0.7, label='Daily Sales')

# 绘制折线图
plt.plot(days, sales, color='blue', marker='o', linestyle='-', label='Trend')

# 添加标题和标签
plt.title('Daily Sales for a Month')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.legend()

# 显示图形
plt.show()
plt.close()
```

#### 1.创建绘图对象

```python
# 1. 用plot.fiture()创建一个绘图对象
# 在一般使用中,我们不会用一个变量来接住这一个对象,如示例程序中,我们全程调用plt.xxx完成了绘图,这并不是一个好习惯,但是比较方便.
# 如果全程使用plt,操作的图就是全局的默认值,这个默认值是最后一个图.
# plt.figure(num="My Custom Figure", figsize=(10, 5), dpi=150, facecolor='lightblue', edgecolor='black', frameon=True, clear=True)
# plt.figure()也用于激活一个已存在的图形.
'''
参数说明:
num:
类型:int | str | Figure | SubFigure | None
说明:指定图形的编号或名称.如果是 int,表示图形的编号；如果是 str,表示图形的名称；如果传入一个 Figure 或 SubFigure 对象，则激活该图形.None 表示创建一个新的图形.
示例:plt.figure(num=1) 或 plt.figure(num="My Figure")

figsize:
类型:tuple[float, float] | None
说明:设置图形的尺寸，以 (宽, 高) 格式,单位为英寸.None 表示使用默认尺寸 (6.4, 4.8).
示例:plt.figure(figsize=(8, 6)) 表示图形宽 8 英寸，高 6 英寸.

dpi:
类型:float | None
说明:图形的分辨率,即每英寸的点数(dots per inch).默认值通常是 100.如果设置为 None,将使用默认的 dpi 值.
示例:plt.figure(dpi=120) 表示分辨率为每英寸 120 点.

facecolor:
类型:ColorType | None
说明:图形背景的颜色.可以使用颜色名(如 'white'）或 RGB 值(如 (0.8, 0.8, 0.8)).None 表示使用默认背景颜色.
示例:plt.figure(facecolor='lightgray')

edgecolor:
类型:ColorType | None
说明:图形边框的颜色，同样可以使用颜色名或 RGB 值.None 表示使用默认边框颜色.
示例:plt.figure(edgecolor='blue')

frameon:
类型:bool
说明:是否绘制图形边框.默认值是 True,表示显示边框.如果设置为 False,则边框将不显示.
示例:plt.figure(frameon=False)

FigureClass:
类型:type[Figure]
说明:用于创建图形的类,默认是 Figure.可以自定义一个 Figure 类，并传入此参数来控制图形对象的创建.
示例:plt.figure(FigureClass=MyCustomFigureClass)

clear:
类型:bool
说明:如果 num 指定了一个已存在的图形,clear=True 会清空该图形的内容.默认为 False,即保留图形内容.
示例:plt.figure(num=1, clear=True) 会清空编号为 1 的图形.

kwargs:
类型:Any
说明:用于传递其他可选的关键字参数,这些参数将传递给 Figure 类的构造函数.
示例:plt.figure(..., tight_layout=True) 将自动调整子图布局.
'''
plt.figure(num="折线图", figsize=(10, 5), dpi=150, facecolor='lightblue', edgecolor='black', frameon=True, clear=True)
```

#### 2.指定具体图类型并导入数据

```python
# 2. 用plt.{chart}( )传入数据和具体要绘制的图标

#折线图
'''
plt.plot() 函数
用途:绘制线图和简单的点图（线图中点之间有连接线）.
常用参数:
x, y:绘制的数据序列,可以是列表、ndarray或 pandas.Series.x 是横坐标,y 是纵坐标.
color:设置线条颜色,颜色可以用英文名、缩写或 RGB 十六进制表示,如 'red' 或 '#FF0000'.
linewidth 或 lw:设置线条宽度,默认值为 1.
linestyle 或 ls:设置线条样式,如 '-'（实线）、'--'（虚线）、'-.'（点划线）、':'（点线）.
marker:设置数据点的标记样式,如 'o'（圆圈）、's'（方形）、'x'（叉形）.
label:为曲线设置标签,便于图例显示.
'''
plt.figure(num="折线图", figsize=(10, 5), dpi=150, facecolor='lightblue', edgecolor='black', frameon=True, clear=True)
plt.plot(days,sales,color='red',linewidth=2,linestyle='--',marker='o',label='Sales')
for d in zip(days,sales):
    plt.plot(d[0],d[1],marker='o',color='black')

plt.show()
plt.close()
```

```python
# 散点图
'''
plt.scatter() 函数
用途:绘制散点图.
常用参数:
x, y:绘制的点的坐标序列,x 为横坐标,y 为纵坐标.
s:设置散点的大小,可以是单个数值或数组.
c:设置点的颜色,可以使用单个颜色或颜色数组.
marker:设置点的形状,如 'o'（圆形）、's'（方形）、'^'（三角形）.
alpha:设置点的透明度,范围在 [0, 1] 之间.
label:为点集设置标签,便于图例显示.
'''
plt.figure(num="散点图", figsize=(10, 5), dpi=150, facecolor='lightblue', edgecolor='black', frameon=True, clear=True)
plt.scatter(days,sales,c='blue',s=8,marker='^',alpha=1,label='Sales')
plt.show()
plt.close()
```

```python
# 柱状图
'''
plt.bar() 函数
用途:绘制条形图,可以是垂直条形图或水平条形图.
常用参数:
x, height:x 为条形图的位置,height 为每个条形的高度（垂直条形图）或宽度（水平条形图）.
width:设置每个条形的宽度,默认值为 0.8.
color:设置条形的颜色,可以是单个颜色或颜色列表.
edgecolor:设置条形边框的颜色.
align:设置条形的对齐方式,'center'（默认）或 'edge'.
label:为条形图设置标签,便于图例显示.
'''
plt.figure(num="直方图", figsize=(10, 5), dpi=150, facecolor='lightblue', edgecolor='black', frameon=True, clear=True)
# 使用 plt.bar() 绘制直方图
plt.bar(
    days,                  # x 轴的位置
    sales,            # 条形的高度
    width=0.6,          # 每个条形的宽度
    color='skyblue',    # 条形的颜色
    edgecolor='black',  # 条形边框的颜色
    align='center',     # 条形对齐方式
    label='Sales'       # 图例标签
)

plt.show()
plt.close()
```

```python
# 直方图
'''
plt.hist() 函数
用途:绘制直方图,用于统计数据的频率分布.
常用参数:
x:需要统计的数据序列.
bins:设置直方图的分组数目,默认为 10.
range:设置数据的范围,如 (min, max).
density:如果为 True,直方图的总面积为 1,适合绘制概率密度图.
color:设置直方图的颜色.
alpha:设置透明度,范围在 [0, 1] 之间.
label:为直方图设置标签,便于图例显示.
'''
# 示例数据
data = np.random.randn(1000)  # 生成一个包含1000个随机数的正态分布数据序列
plt.figure(num="直方图", figsize=(10, 5), dpi=150, facecolor='lightblue', edgecolor='black', frameon=True, clear=True)
# 使用 plt.hist() 绘制直方图
plt.hist(
    data,             # 数据序列
    bins=20,          # 分组数目，划分为20个区间
    range=(-4, 4),    # 数据范围，设定为(-4, 4)
    density=True,     # 如果为 True，直方图的总面积为 1
    color='blue',   # 设置直方图的颜色
    alpha=0.7,        # 设置透明度，0.7 代表70%的不透明度
    label='Frequency' # 图例标签
)

# 添加标题和轴标签
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Density")

# 显示图例
plt.legend()

# 显示图形
plt.show()
plt.close()
```

```python
# 饼图
'''
plt.pie() 函数
用途:绘制饼图,适合显示比例关系.
常用参数:
x:表示各部分大小的序列.
labels:为每个部分设置标签.
colors:设置每个部分的颜色,可以使用颜色列表.
autopct:设置每个部分的百分比显示格式,如 '%.1f%%'.
explode:设置每个部分的偏移量,默认为 0,可以用列表调整特定部分的偏移.
startangle:设置饼图的起始角度,以度数为单位,默认为 0.
shadow:如果为 True,绘制阴影效果.
'''

sizes = [30, 20, 25, 25]  # 各部分大小
labels = ['Category A', 'Category B', 'Category C', 'Category D']  # 各部分标签
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']  # 各部分颜色
plt.figure(num="直方图", figsize=(10, 5), dpi=150, facecolor='lightblue', edgecolor='black', frameon=True, clear=True)
# 使用 plt.pie() 绘制饼图
plt.pie(
    sizes,                # 各部分大小
    labels=labels,        # 标签
    colors=colors,        # 颜色列表
    autopct='%.1f%%',     # 百分比显示格式，保留一位小数
    explode=(0.1, 0, 0, 0), # 偏移量，第一个部分偏移 0.1，其余部分不偏移,偏移就是突出显示
    startangle=90,        # 起始角度，从90度开始绘制
    shadow=True           # 显示阴影效果
)

# 添加标题
plt.title("Pie Chart Example")

# 显示图形
plt.show()
plt.close()

```

```python
# 矩阵图

'''
plt.imshow() 函数
用途:绘制图像或显示矩阵数据,常用于图像处理.
常用参数:
X:输入的图像数据或二维数组.
cmap:设置颜色映射表,如 'gray'、'viridis' 等.
interpolation:设置插值方法,如 'nearest'、'bilinear' 等.
aspect:设置图像的长宽比,'auto'（自动）、'equal'（长宽相等）.
alpha:设置透明度,范围在 [0, 1] 之间.
'''

# 创建一个二维数组数据
data = np.random.rand(10, 10)  # 10x10 随机数矩阵
# 绘制矩阵图
plt.imshow(data, cmap='viridis', interpolation='nearest', aspect='auto', alpha=0.8)

# 添加标题和颜色条
plt.title('matrix')
plt.colorbar(label='size')  # 显示颜色条并设置标签

# 显示图像
plt.show()

'''
只有这种图有特殊的存储函数
'''
plt.imsave('matrix.png', data)
```

#### 3.设置图表信息和布局参数

```python
'''
1. 设置图表标题、标签和图例
plt.title(): 设置图表的标题.
参数:label(标题文本)、fontsize(字体大小)、color(字体颜色)等.

plt.xlabel(): 设置 x 轴的标签.
参数:label(标签文本)、fontsize、color 等.
plt.ylabel(): 设置 y 轴的标签.
参数:label(标签文本)、fontsize、color 等.

plt.legend(): 显示图例，通常与 label 参数配合使用.
参数:loc(图例位置，如 'upper right')、fontsize、frameon(是否显示图例边框)等.

2. 设置图表布局
plt.tight_layout(): 自动调整子图参数，使图表布局更紧凑，避免重叠.
plt.subplots_adjust(): 手动调整子图的边距和间距.
参数:left, right, top, bottom, wspace, hspace 等，用于控制子图之间的间距.

3. 设置网格线
plt.grid(): 显示或隐藏网格线.
参数:True(显示网格)或 False(隐藏网格),color(网格线颜色)、linestyle(网格线样式)等.

4. 设置坐标轴范围
plt.xlim(): 设置 x 轴的范围.
参数:(xmin, xmax).
plt.ylim(): 设置 y 轴的范围.
参数:(ymin, ymax).

5. 设置刻度和刻度标签
plt.xticks(): 设置 x 轴的刻度和刻度标签.
参数:ticks(刻度位置列表)、labels(刻度标签列表)、rotation(标签旋转角度)等.
plt.yticks(): 设置 y 轴的刻度和刻度标签.
参数:与 plt.xticks() 类似.
'''
# 坐标轴操作示例和高级操作在画框的操作和坐标轴的移动部分


plt.figure(num="带图例的折线图", figsize=(10, 5), dpi=150, facecolor=None, edgecolor='black', frameon=True, clear=True)

plt.plot(days,sales,color='red',linewidth=2,linestyle='--',marker='o',label='Sales')



plt.title('Sales Trend',fontsize=16,color='blue')
plt.xlabel('Days',fontsize=12,color='green')
plt.ylabel('Sales',fontsize=12,color='green')
plt.legend()
plt.grid()
plt.xlim(1,30)
plt.xticks([7,14,21,28],['first week','second week','third week','fourth week'],rotation=45)




plt.show()
plt.close()
```

### 画框的操作和坐标轴的移动

```python
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure()
plt.plot(x, y2)
# plot the second curve in this figure with certain parameters
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
# 限制x和y的范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))

# set new ticks
# plt.xticks(): 设置 x 轴的刻度和刻度标签.
# 参数:ticks(刻度位置列表)、labels(刻度标签列表)、rotation(标签旋转角度)等.
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
# set tick labels
plt.yticks([-2, -1.8, -1, 1.22, 3],
           ['$really\ bad$', '$bad$', '$normal$', '$good$', '$really\ good$'])
# to use '$ $' for math text and nice looking, e.g. '$\pi$'

# 通过gca得到当前的边框
# gca = 'get current axis'
ax = plt.gca()
print(ax)
# plt.show()
# 去掉右边框和上边框
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')



# 设置x轴刻度位置一定在底部
ax.xaxis.set_ticks_position('bottom')
# ACCEPTS: [ 'top' | 'bottom' | 'both' | 'default' | 'none' ]


# 把x轴的位置设置在y=0的位置
ax.spines['bottom'].set_position(('data', 0)) 
# ax.spines['bottom'].set_position(('axes', 0)) 
# ax.spines['bottom'].set_position(('outward', 10)) 
'''
set_position(position: tuple[Literal['outward', 'axes', 'data'], float] | Literal['center', 'zero']) -> None
这个方法用于设置坐标轴脊柱的位置，并接受一个参数 position.

position 参数可以是一个元组，也可以是 'center' 或 'zero' 字面量.
当 position 是一个元组时，它由两部分组成:位置类型和数量.
'outward':将脊柱放置在数据区域外，由指定的点数决定.如果是负值，则将脊柱向内放置.
'axes':将脊柱放置在指定的轴坐标上(百分比)，坐标范围是 0 到 1.
'data':将脊柱放置在另一维指定的数据坐标上.
对于快捷方式，以下是一些特殊位置的定义:
'center':将脊柱放置在轴的中心.
'zero':将脊柱放置在数据坐标的原点.
'''


# 设置y轴刻度位置一定在左边
ax.yaxis.set_ticks_position('left')
# ACCEPTS: [ 'left' | 'right' | 'both' | 'default' | 'none' ]

# 把y轴的位置设置在x=0的位置
ax.spines['left'].set_position(('data',0))
plt.show()
```

### 子图创建与设置

```python
# 子图的创建方式(切换法)
plt.figure(figsize=(6, 4))
# plt.subplot(n_rows, n_cols, plot_num)
# 这里的第三个值plot_num是在前两个值作为行列值的矩阵中从上到下，从左到右的顺序进行编号第plot_num个
plt.subplot(2, 2, 1)
plt.plot([0, 1], [0, 1])

plt.subplot(222)
plt.plot([0, 1], [0, 2])

plt.subplot(223)
plt.plot([0, 1], [0, 3])

plt.subplot(224)
plt.plot([0, 1], [0, 4])

plt.tight_layout()
```

```python
# 子图创建的另一种方式：如:
# f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
'''
其中f接受的是一个figure对象,ax1, ax2接受的是两个子图对象
可以在f上设置图总体的属性,如标题、尺寸、布局、背景颜色、边距、保存图形、图形属性、坐标轴、子图、网格、颜色条。
ax{i}上进行具体子图的设置
接受的m*n两个数字表示子图的行列数,比如例子中1*2表示1行2列的子图,ax1表示左侧子图,ax2表示右侧子图。
如果语句是fig, axs = plt.subplots(2, 2, subplot_kw=dict(projection="polar")),axs是一个2*2的数组,每个元素都是一个子图对象,访问各个对象用法是axs[i][j],i表示行,j表示列。
sharex和sharey表示是否共享x轴和y轴,比如例子中sharey=True表示两个子图共享y轴。

'''
# 创建一些测试数据
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# 创建一个画像和子图 -- 图1
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')

# 创建两个子图共享y轴 -- 图2
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)

# 创建四个子图 -- 图4
fig, axs = plt.subplots(2, 2, subplot_kw=dict(projection="polar"))
axs[0, 0].plot(x, y)
axs[1, 1].scatter(x, y)

# 共享 x 轴
plt.subplots(2, 2, sharex='col')

# 共享 y 轴
plt.subplots(2, 2, sharey='row')

# 共享 x 轴和 y 轴
plt.subplots(2, 2, sharex='all', sharey='all')

# 这个也是共享 x 轴和 y 轴
plt.subplots(2, 2, sharex=True, sharey=True)

# 创建标识为 10 的图,已经存在的则删除
fig, ax = plt.subplots(num=10, clear=True)

plt.show()
```
