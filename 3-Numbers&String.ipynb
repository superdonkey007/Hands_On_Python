### **Numbers**

```python
# python的数字包括int,float,bool,complex,其中bool是int的子类，在运算时，True作为1，False作为0处理。
# 很基础的东西(比如加减乘除)不做记录，只记录和其他语言不同的


# 算术运算符

# **	幂运算——返回x的y次幂	a**b 为a的b次方
print(2**3)
# //	取整除——往小的方向取整数
print(9//2)


# 比较运算符

# 浮点数运算有误差，因此不能用==比较，加以阈值限制
a = 0.1 + 0.2
b = 0.3
print(a == b)  # 输出：False
print(abs(a - b) < 1e-9)  # 输出：True


# 赋值运算符
# **=	幂赋值运算符	c **= a 等效于 c = c ** a
a = 10
a **= 2
print(a)
# //=	取整除赋值运算符	c //= a 等效于 c = c // a
a = 9
a //= 2
print(a)
# :=	海象运算符，在表达式中同时进行赋值和返回赋值的值。
string = '123'
print(n := len(string))  # 赋值为3的同时返回3


# 位运算符

# 按位运算符是把数字看作二进制来进行计算的，计算机存储数据使用补码。
# 1.按位与 & 如果两个相应位都为1,则该位的结果为1,否则为0
# 2.按位或 | 只要对应的二个二进位有一个为1时，结果位就为1
# 3.按位异或 ^ 当两对应的二进位相异时，结果为1
# 4.按位取反 ~ 对数据的每个二进制位取反,即把1变为0,把0变为1。~x 类似于 -x-1
# 5.算数左移位 << 运算数的各二进位全部左移若干位，由"<<"右边的数指定移动的位数，高位丢弃，低位补0。
# 6.算数右移位 >> 左边的运算数的各二进位全部右移若干位，对正数：高位补0，低位丢弃。对负数：高位补1，低位丢弃。
a = 60            # 60 = 0011 1100
b = 13            # 13 = 0000 1101
c = 0
c = a & b        # 12 = 0000 1100
print("1 - c 的值为：", c)
c = a | b        # 61 = 0011 1101
print("2 - c 的值为：", c)
c = a ^ b        # 49 = 0011 0001
print("3 - c 的值为：", c)
c = ~a           # -61 = 1100 0011，11000011补码的真值为-61
a = -2             # -2 == 1111 1110
print("4 - c 的值为：", c)
c = a << 2       # -8 = 1111 1000
print("5 - c 的值为：", c)
c = a >> 2       # -1 = 1111 1111
print("6 - c 的值为：", c)


# 逻辑运算符

# x and y
# x or y
# not x


# 成员运算符

# in
# not in


# 身份运算符(内存地址是否相等)

# is 和 is not

# is 是比较对象是否相同(is 表示对象标识符即 object identity)，即用 id() 函数查看的地址是否相同，如果相同则返回 True，如果不同则返回 False。is 不能被重载。
# 由于python的内存优化机制，对于常见变量只要变量值相同，id也一般相同，但对于自定义类就不会
class MyClass:
    def __init__(self, value):
        self.value = value


a = MyClass(10)
b = MyClass(10)
print(a is b)  # 输出：False
# 上述如果实现单例模式，也会返回True


class Singleton:
    _instance = None

    def __new__(cls, *args, **kwargs):
        if not cls._instance:
            cls._instance = super().__new__(cls, *args, **kwargs)
        return cls._instance


a = Singleton()
b = Singleton()
print(a is b)  # 输出：True
# == 是比较两个对象的值是否相等，此操作符内部调用的是 __eq__() 方法。所以 a==b 等效于a.__eq__(b)，所以 = 可以被重载。
```

```python
# 创建和类型转换相关见dataType.ipynb


# 通过序号访问string中的字符
# Python 中的字符串有两种索引方式，从左往右以 0 开始，从右往左以 -1 开始。
# Python 中的字符串不能改变。
# Python 没有单独的字符类型，一个字符就是长度为 1 的字符串。
# 字符串切片 str[start:end]，其中 start（包含）是切片开始的索引，end（不包含）是切片结束的索引。
# 字符串的切片可以加上步长参数 step，语法格式如下：str[start:end:step]
str1='北京理工大学'
print(str1)                 # 输出字符串
print(str1[0:-1])           # 输出第一个到倒数第二个的所有字符
print(str1[0])              # 输出字符串第一个字符
print(str1[2:5])            # 输出从第三个开始到第六个的字符（不包含）
print(str1[2:])             # 输出从第三个开始后的所有字符
print(str1[1:5:2])          # 输出从第二个开始到第五个且每隔一个的字符（步长为2）
print(str1 * 2)             # 输出字符串两次
print(str1 + '你好')         # 连接字符串


# 字符串的运算

# 级联运算，如 "this " "is " "string" 会被自动转换为 this is string。
print("田文镜 ""我上早八")
str2='北京''理工大学'
print(str2)


# 连接和重复运算：字符串可以用 + 运算符连接在一起，用 * 运算符重复。
print("田文镜 "+"我上早八")
print("田文镜 ""我上早八 "*3)

# 成员运算 in 和 not in
print('京' in str2)
print('镜' in str2)


# 转义符 \。
# \r:使用 r 可以让反斜杠不发生转义。r表示raw string
print(r'曾经沧海难为水，\n除却巫山不是云。取次花丛懒回顾，\n半缘修道半缘君。')


# \\:输出\
print('\\')    

# \:单独出现在行尾时表示续行符
print('北京\
理工大学')

# \'和\":单引号和双引号

# \a:响铃？据说执行后电脑有响声。
print('\a')

# \b：退格
print('hello\bworld')

# \000:空
print('\000')

# \n:换行
# \t:横向制表符

# \v：纵向制表符(在命令行里似乎有用)
print("Hello\vWorld!")

# \r:将 \r 后面的内容移到字符串开头，并逐一替换开头部分的字符，直至将 \r 后面的内容完全替换完成。
print('北理工，我上早八\r田文镜')

# \yyy:八进制数，是代表 0~7 的字符，Python 会将其解析为对应的 ASCII 字符。
print("\110\145\154\154\157\40\127\157\162\154\144\41")

# \xyy:十六进制数，以 \x 开头，y 代表的是十六进制字符。
print("\x48\x65\x6c\x6c\x6f\x20\x57\x6f\x72\x6c\x64\x21")

# 字符串格式化——主要内容在Input&Output.ipynb
# 老版方法：%号格式化
name = '田文镜'
age = 40
str3="我叫 %s 今年 %d 岁!" % (name, age)
print ("我叫 %s 今年 %d 岁!" % (name, age))
print(str3)

#新版方法:f或者str.format()
str4=f"我叫 {name} 今年 {age} 岁!"
print(str4)
str5="我叫 {} 今年 {} 岁!".format(name,age)
print(str5)
str6="我叫 {名字} 今年 {年龄} 岁!".format(名字=name,年龄=age)
print(str6)
```
