## **数据类型及数据类型转换**

### **基本数据类型**


```python
'''
Python有六种基本的数据类型,其中三种可修改
不可变数据(3 个):Number(数字)、String(字符串)、Tuple(元组);
可变数据(3 个):List(列表)、Dictionary(字典)、Set(集合)。
'''

'''
数据类型判断  type(a)和isinstance(a,class1)
type()查询变量所指的对象类型,不会认为子类是一种父类类型。 
isinstance(a,class1)判断a是否是class1的一个实例,会认为子类是一种父类类型。
issubclass(class1,class2)
'''
a, b, c, d = 20, 5.5, True, 4+3j
print(type(a), type(b), type(c), type(d))
# bool是int的一个子类
print(issubclass(bool, int))
print(isinstance(a, int))

# 数据的初始化

'''
number的初始化
number包括int、float、bool、complex(复数)。
bool是int的子类,ture可以作为1,false作为0和int作运算
复数可以用a+bj或complex(a,b)定义
Python允许同时为多个变量赋值
'''
a = b = c = 0
print(a, b, c, sep=' ')
print(True+3)
print(False-2)
cp1 = 3+4j
cp2 = complex(3, 4)
print(cp1+cp2)

'''
字符串的初始化
只要为变量分配一个值即可
Python中的字符串是不可变的,C中字符串则可变,可通过序号修改某个值
'''
string = "田文镜,我上早八"
'''
bool类型初始化
只要为变量分配一个值即可

'''
m = True

'''
元组的初始化,用小括号()
'''
tuuple = ('田文镜', '监生', 7, '耀武扬威', '讨债')
# 单个元素的元组,为了区分和数学运算中()的区别,要在元素后加逗号(其他数据类型不用)
tinytuple = ('我上早八',)
# 空元组必须有一对括号
emptytuple = ()
print(type(emptytuple))
print(tuuple)             # 输出完整元组
print(tuuple[0])          # 输出元组的第一个元素
print(tuuple[1:3])        # 输出从第二个元素开始到第三个元素
print(tuuple[2:])
print(tuuple[2::-1])          # 输出从第三个元素开始的所有元素
print(tinytuple * 2)     # 输出两次元组
print(tuuple + tinytuple)  # 连接元组

'''
列表的初始化,用中括号[]
Python中的list不用指定数据类型,这一点和C有很大区别
'''
list = ['田文镜', '监生', 7, '耀武扬威', '讨债']  # 定义一个列表
tinylist = ['我上早八']
emptylist = []
print(type(emptylist))
print(list)            # 打印整个列表
print(list[0])         # 打印列表的第一个元素
print(list[1:3])       # 打印列表第二到第四个元素(不包含第四个元素)
print(list[2:])        # 打印列表从第三个元素开始到末尾
print(tinylist * 2)    # 打印tinylist列表两次
print(list + tinylist)  # 打印两个列表拼接在一起的结果

'''
集合的初始化,用大括号{}或set()
Python 中的集合(Set)是一种无序、可变的数据类型,用于存储唯一的元素。
集合中的元素不会重复,并且可以进行交集、并集、差集等常见的集合操作。
'''
sets = {'田文镜', '监生', 7, '耀武扬威', '讨债', 7}
tinysets = {'我上早八', '田文镜'}
# 空集合用set()创建,以此和字典做区分
emptyset = set()
print(type(emptyset))
print(sets - tinysets)  # 差集
print(sets | tinysets)  # 并集
print(sets & tinysets)  # 交
print(sets ^ tinysets)  # 不同时存在的元素
print(sets)   # 输出集合,重复的元素被自动去掉

'''
字典的初始化,用大括号{}
字典是无序的对象集合,字典当中的元素是通过键来存取的,而不是通过偏移存取。
字典是一种映射类型,字典用 {} 标识,它是一个无序的 键(key) : 值(value) 的集合。
键(key)必须使用不可变类型。
'''
dic = {}
print(type(dict))
dic['田文镜'] = "我上早八"
dic['北京理工大学'] = "我也上早八"

tinydict = {'name': 'ckp', 'code': '3220241343'}


print(dic['田文镜'])       # 输出键为 '田文镜' 的值
print(dic['北京理工大学'])           # 输出键为 '北京理工大学' 的值
print(tinydict)          # 输出完整的字典
print(tinydict.keys())   # 输出所有键
print(tinydict.values())  # 输出所有值
'''
在 Python3 中,bytes 类型表示的是不可变的二进制序列(byte sequence)。
与字符串类型不同的是,bytes 类型中的元素是整数值(0 到 255 之间的整数),而不是 Unicode 字符。
bytes 类型通常用于处理二进制数据,比如图像文件、音频文件、视频文件等等。在网络编程中,也经常使用 bytes 类型来传输二进制数据。
创建 bytes 对象的方式有多种,最常见的方式是使用 b 前缀:x = b"hello"
此外,也可以使用 bytes() 函数将其他类型的对象转换为 bytes 类型。
bytes() 函数的第一个参数是要转换的对象,第二个参数是编码方式,如果省略第二个参数,则默认使用 UTF-8 编码
'''
x = b"hello"
y = x[1:3]  # 切片操作,得到 b"el"
z = x + b"world"  # 拼接操作,得到 b"helloworld"
print(y, z)
x = bytes("hello", encoding="utf-8")
print(x)
# ord函数将字母转换为相应的整数值
print(x[0], ord('h'))
```

### **数据类型转换**

```python
# 隐式类型转换
# 1.数学计算中int转float，bool转int
num_f = 3.14
print(type(num_f))
num_i = 3
print(type(num_i))
sum = num_f+num_i  # int转为float
print(type(sum))
print('{0:3.2f}'.format(sum))
sum = sum+True  # True转为1，1再转为float
print('{0:3.2f}'.format(sum))
# 2.布尔运算:在布尔运算and, or, 和 not中会导致int->bool,float->bool的隐式类型转换。
result = 42 and True  # result = True，因为非零数字转换为True
print(result)
result = 3.5 and True  # result = True，因为非零数字转换为True
print(result)
result = 0 or True   # result = 0，
print(result)
result = -1.5 and True   # result = 0，
print(result)
```
